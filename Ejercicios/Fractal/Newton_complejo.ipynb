```julia
# creamos una malla de números complejos, uso ; para evitar 
iniciales = [x + y*im for x in -2:0.001:2, y in -2:0.001:2];
```

```julia
f(x) = x - (x^4+4)/(4*x^3)
```

```julia
iterator = deepcopy(iniciales)
[iterator .= f.(iterator) for i in 1:100]
iterator
```

```julia
# Primero un vecort que contenga las raices.
raices = [x+im*y for x in -1:2:1, y in -1:2:1]
```

```julia
# Ahora creamos una función que nos identifique las raices y asigne un cero cuando los valores no convergen.
function convert_root(root)
    tolerance = 1e-6
    if abs(root-raices[1]) < tolerance
        return 1
    elseif abs(root-raices[2]) < tolerance
        return 2
    elseif abs(root-raices[3]) < tolerance
        return 3
    elseif abs(root-raices[4]) < tolerance
        return 4
    else
        return 0
    end
end
```

```julia
# Aplicamos la función al arreglo
toplot = convert_root.(iterator);
```

```julia

using Plots

#heatmap
heatmap(toplot, title="Fractal de Newton f(x)=x^4+4", c=:viridis, aspect_ratio=1, color=:grays, legend=false)
```

```julia
savefig("fractal_newton_fuction_f.png")
```

```julia
function f!(x::Matrix{Complex{Float64}})
    x.= f.(x) 
end
```

```julia
iterator = deepcopy(iniciales);
for i in 1:100
    f!(iterator)
end
iterator
```

```julia
heatmap(toplot, title="Fractal de Newton f(x)=x^4+4", c=:viridis)
```

```julia
savefig("fractal_newton_fuction_f1.png")
```

## Ejemplo  2
$$x^3+1=0$$

```julia
# creamos una malla de números complejos, uso ; para evitar 
iniciales = [x + y*im for x in -2:0.001:2, y in -2:0.001:2];

h(x) = x - (x^3+1)/(3*x^2)

iterator = deepcopy(iniciales)
[iterator .= h.(iterator) for i in 1:100]
iterator
```

```julia
# Primero un vecort que contenga las raices.
raices = [-1, 1/2+im*sqrt(3)/2, 1/2-im*sqrt(3)/2]
```

```julia
# Ahora creamos una función que nos identifique las raices y asigne un cero cuando los valores no convergen.
function convert_root(root)
    tolerance = 1e-6
    if abs(root-raices[1]) < tolerance
        return 1
    elseif abs(root-raices[2]) < tolerance
        return 2
    elseif abs(root-raices[3]) < tolerance
        return 3
    else
        return 0
    end
end
```

```julia
# Aplicamos la función al arreglo
toplot = convert_root.(iterator);
```

```julia
using Plots

#heatmap
heatmap(toplot, title="Fractal de Newton h(x)=x^3+1", c=:viridis, aspect_ratio=1, color=:blues, legend=false)
```

```julia
savefig("fractal_newton_fuction_h1.png")
```

```julia
function h!(x::Matrix{Complex{Float64}})
    x.= h.(x) 
end

iterator = deepcopy(iniciales);
for i in 1:100
    h!(iterator)
end
iterator
```

```julia
heatmap(toplot, title="Fractal de Newton h(x)=x^3+1", c=:viridis)
```
